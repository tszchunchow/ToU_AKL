```python
from PIL import Image
import os
import glob
from utils import *
```

```python
def year(scen_code):
    if scen_code == 12:
        return 18
    else:
        return YEAR

def crop_img(img_path, crop_box):
    with Image.open(img_path) as img:
        cropped_img = img.crop(crop_box)

        # Save the cropped image to the output directory
        output_fn = os.path.basename(img_path).split("_014")[0] + '.jpg'
        output_path = os.path.join(output_dir, output_fn)
        cropped_img.save(output_path)
        print(f'Cropped and saved {output_fn}')


# Define the directory paths
INPUT_DIR = '/Users/tszchun.chow/Library/CloudStorage/OneDrive-SharedLibraries-AucklandTransport/Congestion Charging - General/3. Policy & Strategy/Scheme analysis/Modelling outputs/001_Plots/2026/GC_and_TT_diff_Plots'
output_dir = '../outputs/images/GC'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)
# Define the crop box (left, upper, right, lower)
# Adjust these values based on the area you want to crop
left = 173
upper = 123

crop_box = (left, upper, left+3159, upper+2260)

search_location = INPUT_DIR + '/*/*/*.jpg'
for period in PERIOD_MAP:
    period_code = PERIOD_MAP[period]
    for scenario in SCENARIO_MAP:
        scen_code = SCENARIO_MAP[scenario]
        if scen_code not in [12, 16, 76]:
            model_code = str(year(scen_code)) + str(period_code) + str(scen_code)
            for img_path in glob.glob(search_location):
                if (model_code in img_path) and ('GC' in os.path.basename(img_path)) and img_path.endswith('Region_Ore_Dru.jpg'):
                    print(f'Opening {img_path}')
                    crop_img(img_path, crop_box)
```

```python
# Iterate over each file in the input directory
for filename in os.listdir(search_location):
    if filename.endswith('Region_Ore_Dru.jpg'):  # filter image files
        image_path = os.path.join(search_location, filename)
        
        # Open and crop the image
        with Image.open(image_path) as img:
            cropped_img = img.crop(crop_box)
            
            # Save the cropped image to the output directory
            output_fn = filename.split("_014")[0] + '.jpg'
            output_path = os.path.join(output_dir, output_fn)
            cropped_img.save(output_path)
            print(f'Cropped and saved {filename}')
```
